```python
columns1 = ['Kivisüsi, tuhat t', 'Põlevkivi, tuhat t', 'Tükkturvas, tuhat t',
       'Turbabrikett, tuhat t', 'Küttepuud, tuhat tm', 'Puiduhake ja -jäätmed, tuhat tm',       
       'Maagaas, mln m³', 'Raske kütteõli, tuhat t', 'Põlevkiviõli (raske fraktsioon), tuhat t',
       'Kerge kütteõli ja diislikütus, tuhat t', 'Diislikütus, tuhat t',
       'Autobensiin, tuhat t', 'Elektrienergia, GWh', 'Soojus, GWh']
```

```python
df = pd.DataFrame()
for col in columns1:
    df0 = df1[['Aasta', 'Tegevusala', col]]
    if col == 'Kerge kütteõli ja diislikütus, tuhat t': df0['Liik'] = 'Kerge kütteõli, tuhat t'
    elif col == 'Põlevkiviõli (raske fraktsioon), tuhat t': df0['Liik'] = 'Põlevkiviõli, tuhat t'
    else: df0['Liik'] = col
    df0.columns = ['Aasta', 'Tegevusala', 'Kogus', 'Liik']
    df = df.append(df0)
```

```python
df['Ühik'] = df.apply(lambda row: str(row['Liik']).split(',')[-1].strip(),axis=1)
df['Liik'] = df.apply(lambda row: str(row['Liik']).split(',')[0].strip(),axis=1)
df.head()
```

```python
df = df.drop(index = df[df['Kogus'] == '..'].index,axis=0)
```

```python
df['Kogus'] = df['Kogus'].astype(float)
df.dtypes
```
